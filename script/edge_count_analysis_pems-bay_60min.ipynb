```python
%matplotlib widget
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.sparse as sp
import csv
from collections import Counter
from torch_geometric.utils import from_scipy_sparse_matrix
```

```python
data_folder = "../logs/2025-07-01_10-57-17_pems-bay_pred-15min_his-60min_ray-semi-dec-server-free-fl-distance_online-training_with-distribution-adaptive-edge-score/edge_counts/cloudlet_0.csv"

# Read the CSV file
with open(data_folder, newline='') as csvfile:
    reader = csv.reader(csvfile)
    headers = next(reader)  # Edge IDs (not used here directly)
    values = next(reader)   # Selection counts as strings

# Convert values to integers
selection_counts = list(map(int, values))

# Count how many times each selection count appears
count_distribution = Counter(selection_counts)

# Sort by selection count (from high to low)
for count in sorted(count_distribution.keys(), reverse=True):
    print(f"{count} times selected: {count_distribution[count]} edges")

# Calculate and print the average number of times an edge was selected
average_selection = sum(selection_counts) / len(selection_counts)
print(f"\nAverage number of times an edge was selected: {average_selection:.2f}")
```
